```python
import pandas as pd
import numpy as np
```

```python
file_path = "Inventory.xlsx"

component = pd.read_excel(file_path, sheet_name="Component_Master")
bom = pd.read_excel(file_path, sheet_name="BOM_Mapping")
```

```python
component.head()
```

```python
bom.head()
```

```python
component.columns = component.columns.str.strip().str.replace(" ", "_")
bom.columns = bom.columns.str.strip().str.replace(" ", "_")
```

```python
component.describe()
```

```python
bom.describe()
```

```python
print(component.isnull().sum())
```

```python
print(bom.isnull().sum())
```

```python
component.fillna(0,inplace=True)
```

```python
print(component.isnull().sum())
```

```python
component = component.dropna(how="all")
bom = bom.dropna(how="all")
```

```python
component["Component_Name"] = component["Component_Name"].str.strip()
component["Category"] = component["Category"].str.strip()


bom["Device_Name"] = bom["Device_Name"].str.strip()
bom["Component_Name"] = bom["Component_Name"].str.strip()
```

```python
component = component.dropna(subset=["Component_Name"])
bom = bom.dropna(subset=["Component_Name"])
```

```python
component["Stock_Qty"] = component["Stock_Qty"].fillna(0)
component["Reorder_Level"] = component["Reorder_Level"].fillna(0)
component["Unit_Cost_INR"] = component["Unit_Cost_INR"].fillna(0)
```

```python
component["Inventory_Value"] = component["Stock_Qty"] * component["Unit_Cost_INR"]
```

```python
component["Stock_Status"] = component.apply(
    lambda row: "Low Stock"
    if row["Stock_Qty"] <= row["Reorder_Level"]
    else "OK",
    axis=1)
```

```python
with pd.ExcelWriter("Inventory_Cleaned.xlsx") as writer:
    component.to_excel(writer, sheet_name="Component_Master", index=False)
    bom.to_excel(writer, sheet_name="BOM_Mapping", index=False)

print("Cleaning completed. New file created: Inventory_Cleaned.xlsx")
```
